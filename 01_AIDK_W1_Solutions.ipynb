## Python Basics - Functions

`def` lets you create your own reusable function. Parameters go inside the brackets.

````python
def calculate_average(numbers):
    total = sum(numbers)
    average = total / len(numbers)
    return average

scores = [72, 85, 60, 91]
print(calculate_average(scores))
````

````python
def check_pass(score):
    if score >= 60:
        return "Pass"
    else:
        return "Try again"

print(check_pass(72))
````

## Practice 1 - Complete a Class Score Summary

Fill every blank in the code to calculate the class average and assign the correct status.

Expected output:


````python
scores = [72, 85, 60, 91, 45]

total_score = sum(scores)
number_of_scores = len(scores)
average_score = total_score / number_of_scores

if average_score >= 80:
    status = "Excellent"
elif average_score >= 60:
    status = "Pass"
else:
    status = "Needs support"

print("Average:", average_score)
print("Status:", status)
````

## Practice 2 - Write a Result Loop

Starting with the list below, write code that loops through every score and prints its result:

- `Excellent` for 80 or higher
- `Pass` for 60 or higher
- `Needs support` otherwise

Expected output:

```text
72 Pass
85 Excellent
60 Pass
91 Excellent
45 Needs support
```

````python
scores = [72, 85, 60, 91, 45]

for score in scores:
    if score >= 80:
        status = "Excellent"
    elif score >= 60:
        status = "Pass"
    else:
        status = "Needs support"

    print(score, status)
````

## Practice 3 - Build a Student Summary

Write a function called `summarise_student(student)` that:

1. Calculates the average of the scores in the dictionary.
2. Classifies the result as `Excellent` for 80 or higher, `Pass` for 60 or higher, or `Needs support` otherwise.
3. Prints the student's name, average, and status.

Expected output:

```text
Name: Aisha
Average: 84.0
Status: Excellent
```

````python
student = {
    "name": "Aisha",
    "scores": [88, 73, 91]
}

def summarise_student(student):
    average = sum(student["scores"]) / len(student["scores"])

    if average >= 80:
        status = "Excellent"
    elif average >= 60:
        status = "Pass"
    else:
        status = "Needs support"

    print("Name:", student["name"])
    print("Average:", average)
    print("Status:", status)

summarise_student(student)
````
